```python
import os

import pandas as pd
from matplotlib import pyplot as plt
```

```python
cqm_schedules_df = pd.read_csv(os.path.join("experiments", "cqm_scheduling", "cqm_limits", "scheduled_instances.csv"))
```

```python
cqm_schedules_df = cqm_schedules_df[["variables_count", "constraints_count", "cost_increase"]]
```

```python
fig,ax = plt.subplots()

sol_found = cqm_schedules_df[cqm_schedules_df['cost_increase'].notnull()]
no_sol = cqm_schedules_df[cqm_schedules_df['cost_increase'].isnull()]

ax.scatter(sol_found["constraints_count"], sol_found["variables_count"], color='green', label='solution found')
ax.scatter(no_sol["constraints_count"], no_sol["variables_count"], color='red', label='no solution')
ax.legend(prop={'size': 12})
ax.set_xlabel("Number of constrains", fontsize=14)
ax.set_ylabel("Number of variables", fontsize=14)
ax.set_title("Results of workflow scheduling on the hybrid CQM solver", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=12)

fig.savefig(os.path.join("experiments", "cqm_scheduling", "cqm_limits", "cqm_limits.png"), bbox_inches='tight')
```
